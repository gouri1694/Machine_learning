**Probability** <br>
- The likelihood of the occurrence of an event is known as probability. <br>
- The value of probability of an event lies between 0 and 1. <br>


## Types of Events <br>

#### Simple Event <br>

- Any event containing a single element of a sample space. <br>

#### Compound Event <br>

- Any event containing two or more elements of a sample space. <br>

#### Dependent Event <br>

- If the occurrence of an event is influenced by another event, it is called a 'Dependent Event'. <br>

#### Independent Event <br>

- If the occurrence of an event is not influenced by another event, it is called an 'Independent Event'. <br>

#### Exhaustive Events <br>

- A set of events devouring the entire sample space. <br>

#### Mutually Exclusive <br>

- Two events are said to be mutually exclusive events when both cannot occur at the same time. <br>

### Properties of Probability <br>

#### Property 1 <br>

- Probability value lies between 0 and 1. <br>

- **0Probability1** <br>

#### Property 2 <br>

- The probability of an impossible event is 0. <br>

- **P()=0** <br>

#### Property 3 <br>

- The probability of a confirmed event is 1. <br>

- **P(S)=1** <br>

#### Property 4 <br>

- The sum of the probabilities of an event, and its complementary is 1. <br>

- **P(A)+P(A)=1** <br>


### Addition Rule <br>

#### Mutually Exclusive Events <br>

- If A and B are mutually exclusive events, <br>

- P(AB)=P(A)+P(B) <br>

#### Non-mutually Exclusive Events <br>

- If A and B are non-mutually exclusive events, <br>

- P(AB)=P(A)+P(B)P(AB) <br>


### Multiplication Rule <br>

#### Independent Events <br>

- If A and B are two Independent Events, <br>

- P(AB)=P(A)P(B) <br>

#### Dependent Events <br>

- If A and B are two Dependent Events, <br>

- P(AB)=P(A)P(B/A) <br>

### Central Limit Theorem <br>

Suppose there is a random variable X, CLT states that, it may or may not follow Gaussian distribution with some mean and standard deviation, but when sample size>30 is taken with multiple samples, it follows Gaussian distribution. <br>
 <br>

The CLT states that regardless of the shape of *population distribution*, the distribution of *sample means* will be approximately *normal*.<br>
or<br>
The Central Limit Theorem (CLT) states that the sampling distribution of a sample mean approaches a normal distribution, as the sample size gets larger.<br>
    
Consider a population with<br>
- Mean <br>
- Standard Deviation <br>

take sufficiently large random samples from the population with replacement, then the distribution of the sample mean will be approximately normally distributed.<br>

This will hold true regardless of whether the source population is normal or skewed, provided the sample size is sufficiently large (usually n > 30).<br>

- The distribution of sample mean will become more normal as sample size increases <br>
- Good rule thumb: Sample distribution will be approximately normal if their sample size is n>=30 <br>

 <br>

Positive skew ex: Wealth distribution, length of comments in youtube/social media <br>
Mode< Median< Mean <br>

Symmetrical distribution ex: Age, weight, height,IRIS <br>
Mean< Median< Mode <br>

Negative Skew ex: Life span of Human being <br>
Mean< Median < Mode <br>

**Covariance** <br>
- Quantify the relationship between x & y <br>
- Measure of the joint variability of two random variables and describes the relationship between these two variables <br>
- It is defined as the expected value of product of two random variables deviations from their means. <br>
 <br>

 <br>

- It can take positive or negative or 0 value <br>
- +ve value indicate --> 2 random variable tend to vary in same direction <br>
- -ve value indicate --> 2 random variable tend to vary in opposite direction <br> <br>
- 0 value indicate --> 2 random variable don't vary together <br>

**Disadvantage:** <br>
- With Covariance we are able to quantify the relationship, but how much directly correlated is not known. It is not limited (how much positive or how much negative). No limited range is present. <br>

```python
x = np.array([1,3,5,6])
y = np.array([-2,-4,-5,-6])#this will return the covariance matrix of x,y containing x_variance, y_variance on diagonal elements and covariance of x,y
cov_xy = np.cov(x,y)
print(cov_xy)
```

If covariance value is positive: <br>
- When x increases, y increases <br>
- When x decreases, y decreases <br>

```python
from scipy.stats import powerlognorm
import numpy as np
numargs=powerlognorm.numargs
a,b=4.32,3.18
rv=powerlognorm(a,b)
print('RV : \n',rv)
quantile=np.arange(0.01,1,0.1)
#Random variates
R=powerlognorm.rvs(a,b)
print('Random variates:\n',R)
#PDF
R1=powerlognorm.pdf(a,b,quantile)
print('\nProbability Distribution:',R1)
sns.displot(R1,kind='kde')
```
